```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sktime.transformations.panel.rocket import MiniRocket
import warnings
warnings.filterwarnings('ignore')
```

```python
merged_df = behavior.merge(output_stress,how = 'inner', on = ['date','user_id'])
merged_df.info()
```

```python
# Identify feature and target columns
exclude_cols = ['user_id', 'date', 'stress_category']
feature_cols = [col for col in merged_df.columns if col not in exclude_cols]
target_cols = ['stress_category']

print(f"Features: {len(feature_cols)} columns")
print(f"Targets: {target_cols}")
print(f"Merged data shape: {merged_df.shape}")
```

```python
def reshape_to_3d(df, feature_cols, target_cols, min_days=14):

    df = df.sort_values(['user_id', 'date'])
    users = df['user_id'].unique()

    X_list = []
    y_list = []
    date_list = []
    valid_users = []

    for user in users:

        user_df = df[df['user_id'] == user]

        # Skip users with insufficient data
        if len(user_df) < min_days:
            continue

        features = user_df[feature_cols].values.T   # (features, days)
        targets = user_df[target_cols].values       # (days, 1)
        dates = user_df['date'].values              # (days,)

        X_list.append(features)
        y_list.append(targets)
        date_list.append(dates)
        valid_users.append(user)

    return X_list, y_list, valid_users, date_list
```

```python
X, y, user_ids, date_list = reshape_to_3d(
    merged_df, feature_cols, target_cols
)
```

```python
def create_sliding_windows(X_list, y_list, users, dates, window_size=10):

    windows_X = []
    windows_y = []
    meta_user = []
    meta_date = []

    # Iterate per user
    for user_idx, features in enumerate(X_list):

        targets = y_list[user_idx]
        user_dates = dates[user_idx]

        n_days = features.shape[1]  # days for THIS user

        # Skip users that still don't have enough data
        if n_days < window_size:
            continue

        for day in range(window_size, n_days):

            # Extract rolling window
            window = features[:, day-window_size:day]
            target = targets[day]

            windows_X.append(window)
            windows_y.append(target)

            # Save metadata
            meta_user.append(users[user_idx])
            meta_date.append(user_dates[day])

    # Convert to numpy arrays
    X_windows = np.stack(windows_X)
    y_windows = np.stack(windows_y)

    # Convert to (samples, timesteps, features)
    X_windows = np.transpose(X_windows, (0, 2, 1))

    metadata = pd.DataFrame({
        "user_id": meta_user,
        "date": meta_date
    })

    return X_windows, y_windows, metadata
```

```python
X_windows, y_windows, metadata = create_sliding_windows(X, y, user_ids, date_list, window_size=10)
```

```python
from sklearn.model_selection import GroupShuffleSplit

gss = GroupShuffleSplit(test_size=0.2, random_state=40)

train_idx, test_idx = next(
    gss.split(X_windows, y_windows, groups=metadata['user_id'])
)

X_train = X_windows[train_idx]
X_test = X_windows[test_idx]

y_train = y_windows[train_idx]
y_test = y_windows[test_idx]

meta_train = metadata.iloc[train_idx]
meta_test = metadata.iloc[test_idx]
```

```python
metadata.user_id.unique()
```

```python
# Check available parameters
print("MiniRocket parameters:", MiniRocket().get_params().keys())
```

```python

# Phase 1:MiniRocket Transformation

from sktime.transformations.panel.rocket import MiniRocket


num_kernels = 200
max_dilations_per_kernel = 9
window_size = 10

minirocket = MiniRocket(
    num_kernels=num_kernels,
    max_dilations_per_kernel=max_dilations_per_kernel,
    random_state=42,
    n_jobs=-1
)

minirocket.fit(X_train)           # X_train should be (n_samples, timesteps, features)
X_train_mr = minirocket.transform(X_train)
X_test_mr = minirocket.transform(X_test)

print("MiniRocket outputs:", X_train_mr.shape, X_test_mr.shape)




print(f"Original training shape: {X_train.shape}")
print(f"MiniRocket training features: {X_train_mr.shape}")
print(f"MiniRocket test features: {X_test_mr.shape}")
```

```python

# FINAL FEATURE SELECTION
import numpy as np
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import mutual_info_classif
from sklearn.ensemble import RandomForestClassifier


def select_features_minirocket(
    X_train, X_test, y_train,
    stage1_k=1100,
    final_k=900,
    use_mutual_info=False,
    random_state=42
):

    X_train_np = np.asarray(X_train)
    X_test_np = np.asarray(X_test)
    y_train_np = np.asarray(y_train).ravel()   # ensure 1D target

    n_features = X_train_np.shape[1]

    # Clamp to available features
    stage1_k = min(stage1_k, n_features)
    final_k = min(final_k, stage1_k)

    # Stage 1: FILTER SELECTION (UNIVARIATE)
    if use_mutual_info:

        scores = mutual_info_classif(
            X_train_np,
            y_train_np,
            discrete_features=False,
            random_state=random_state
        )

    else:
        scores, _ = f_classif(X_train_np, y_train_np)

    scores = np.nan_to_num(scores)

    stage1_idx = np.argsort(scores)[-stage1_k:]

    X_train_f = X_train_np[:, stage1_idx]
    X_test_f = X_test_np[:, stage1_idx]


    # Stage 2: RANDOM FOREST IMPORTANCE
    rf = RandomForestClassifier(
        n_estimators=200,
        max_depth=9,
        min_samples_leaf=2,
        max_features=0.6,
        random_state=random_state,
        n_jobs=-1,
        class_weight="balanced"   # important for stress imbalance
    )

    rf.fit(X_train_f, y_train_np)

    importances = rf.feature_importances_

    stage2_idx = np.argsort(importances)[-final_k:]

    X_train_sel = X_train_f[:, stage2_idx]
    X_test_sel = X_test_f[:, stage2_idx]

    selected_indices = stage1_idx[stage2_idx]

    return X_train_sel, X_test_sel, selected_indices
```

```python
# FEATURE SELECTION (FINAL)

X_train_selected, X_test_selected, selected_indices = select_features_minirocket(
    X_train_mr, X_test_mr, y_train, stage1_k=1200, final_k=1000, use_mutual_info=False
)
```

# Training

```python
Phase 3: Train & Evaluate Individual Models

from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

print("=== TRAINING INDIVIDUAL MODELS ===\n")
# Convert y to numpy (1D)
y_train_np = np.array(y_train).ravel()
y_test_np = np.array(y_test).ravel()
num_positive = np.sum(y_train_np == 1)   # Stressed
num_negative = np.sum(y_train_np == 0)   # Good

scale_pos_weight = num_negative / num_positive

print("scale_pos_weight:", scale_pos_weight)

# Define models with optimized parameters
models = {
    # 'RandomForest': RandomForestClassifier(
    #     n_estimators=200,
    #     max_depth=8,
    #     min_samples_split=8,
    #     min_samples_leaf=4,
    #     max_features=0.6,
    #     random_state=42,
    #     n_jobs=-1,
    #     class_weight=None
    # ),

    # 'GradientBoosting': GradientBoostingClassifier(
    #     n_estimators=200,
    #     max_depth=3,
    #     learning_rate=0.05,
    #     subsample=0.8,
    #     min_samples_split=10,
    #     random_state=42
    # ),

    'XGBoost': XGBClassifier(
        n_estimators=250,
        max_depth=2,
        learning_rate=0.03,
        subsample=0.75,
        colsample_bytree=0.75,
        random_state=42,
        n_jobs=-1,
        reg_alpha=1,
        reg_lambda=2,
        tree_method='auto',
        eval_metric='logloss',
        scale_pos_weight=scale_pos_weight
    )
}


# ------------------------------------------------
# Store results
# ------------------------------------------------
model_results = {}
predictions = {}

for name, model in models.items():
    prob=[0.35]
    print(f"Training {name}...")
    for i in prob:
      print(f'for threshold = {i}')
      # Fit model
      model.fit(X_train_selected, y_train_np)

      # Predictions
      y_pred_train = model.predict(X_train_selected)

      # y_pred_test = model.predict(X_test_selected)
      probs = model.predict_proba(X_test_selected)[:,1]
      y_pred_test = (probs > i).astype(int)

      # Accuracy
      train_acc = accuracy_score(y_train_np, y_pred_train)
      test_acc = accuracy_score(y_test_np, y_pred_test)

      # Save results
      model_results[name] = {
          'model': model,
          'train_accuracy': train_acc,
          'test_accuracy': test_acc
      }

      predictions[name] = y_pred_test

      # ------------------------------------------------
      # Print results
      # ------------------------------------------------
      print(f"Train Accuracy: {train_acc:.4f}")
      print(f"Test Accuracy:  {test_acc:.4f}\n")

      print("Classification Report (Test):")
      print(classification_report(
          y_test_np,
          y_pred_test,
          target_names=["Stressed", "Good"]
      ))
      print("="*50)
```
